```python
df = df[16150:]
```

```python
df.loc[:,'datetime'] = pd.to_datetime(df[['year','month','day','hour']])
df.head()
```

```python
df = df.drop(['year', 'month', 'day', 'hour'], axis=1)
df.head()
```

```python
df = df.fillna(df.groupby(df.datetime.dt.hour).transform('mean'))
```

```python
fig, ax = plt.subplots(figsize=(13,6))

ax.plot(df['CO'])
ax.set_xlabel('Date')

fig.autofmt_xdate()
plt.tight_layout()
```
